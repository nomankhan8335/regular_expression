# Creating Series Object from List

```python
# string
country = ['India','Pakistan','USA','Nepal','Srilanka']

pd.Series(country)
```

```python
runs = [13,24,56,78,100]

runs_ser = pd.Series(runs)
runs_ser
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
```

```python
# setting a name
marks = pd.Series(marks,index=subjects,name="Harsh's marks")
marks
```

# Creating Series Object from a Dictionary

```python
marks_1 = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks_1,name='nitish ke marks')
marks_series
```

```python
marks_2 = {
    'maths':67,
    'english':57,
    'science':np.nan,
    'hindi':100
}

marks_series_2 = pd.Series(marks_2,name='nitish ke marks')
marks_series_2
```
